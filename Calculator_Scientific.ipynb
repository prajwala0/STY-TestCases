```python
import math

def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def mod(a,b):
    return a%b
def square_root(a):
    return math.sqrt(a)
def sine(a):
    return math.sin(math.radians(a))
def cosine(a):
    return math.cos(math.radians(a))
def tan(a):
    return math.tan(math.radians(a))


while True:
    choice = input('Select number for choice of Operation : \n 1.Addition \n 2.Subraction \n 3.Multiplication \n 4.Division \n 5.Modulus \n 6.SquareRoot \n 7.sine \n 8.cosine \n 9.tan \n')
    
    if choice in ('1','2','3','4','5') :
        a = int(input('Enter first number : '))
        b = int(input('Enter second number : '))

        if choice=='1':
            print(a,'+',b,'=', add(a,b))
        elif choice=='2':
            print(a,'-',b,'=', sub(a,b))
        elif choice=='3':
            print(a,'*',b,'=', mul(a,b))
        elif choice=='4':
            print(a,'/',b,'=', div(a,b))
        elif choice=='5':
            print(a,'%',b,'=', mod(a,b))
            
    elif choice in('6','7','8','9'):
        a=int(input('Enter the value: '))
        if choice =='6':
            sqrt=square_root(a)
            print(a,'**(1/2)= ',sqrt)
        elif choice =='7':
            s = sine(a)
            print('sin',a,'= ',s)
        elif choice =='8':
            c = cosine(a)
            print('cos',a,'= ',c)
        elif choice =='9':
            t = tan(a)
            print('tan',a,'= ',t)
    else:
        print('Invalid Operation')
        
    next_cal = input('If you want to continue calculation? (y/n) :')
    if next_cal == 'n':
        print('Calcualtion will stop')
        break
```
